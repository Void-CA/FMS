```python
import pandas as pd 
import numpy as np 
import plotly.express as px
```

```python
fms = pd.read_csv('../data/FMS.csv')
fms["year"] = fms["year"].replace({"2023A":"2023", "2023B":"2023"})
fms = fms.sort_values(by="year")
fms.head()
```

## Visualización de los Datos (Exploración de Datos)

```python
scale_columns = fms.select_dtypes(include=[np.number]).columns[1:]
fms_scaled = fms.copy()
fms_scaled[scale_columns] = fms_scaled[scale_columns].groupby(fms_scaled["year"]).transform(lambda x: (x - x.mean()) / x.std())
```

```python
fig = px.box(fms_scaled, x="year", y="PTB", color="country", title="PTB by Country and Year")
fig.show()
```

```python
filtered_year = fms[fms["year"] == "2023"]
filtered_year = filtered_year.sort_values(by="PTB")
px.bar(filtered_year, x="country", y="PTB", title="PTB by Country in 2023", color="country")
```

```python
fms_year = fms_scaled[fms_scaled["year"] == "2023"]
px.scatter(fms_year, x="PTB", y="PTS", color="country", title="PTS vs PTB by Country")
```
